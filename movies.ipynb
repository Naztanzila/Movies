```python
tags_df = pd.read_csv("/kaggle/input/movies-data/tags.csv")
```

```python
links_df=pd.read_csv("/kaggle/input/movies-data/links.csv")
```

```python
matrix_movie_id = movie[movie['title'] == 'Matrix, The (1999)']['movieId'].values[0]

# Merging
merged_df = pd.merge(tags_df, links_df, on='movieId')

# Filtering the merged DataFrame for the specified movieId
filtered_df = merged_df[merged_df['movieId'] == matrix_movie_id]

# Extracting the tags for the specified movie
tags_for_matrix = filtered_df['tag'].tolist()

print("Tags for 'Matrix, The (1999)':", tags_for_matrix)
```

```python
import matplotlib.pyplot as plt
```

```python
movie_title = 'Fight Club (1999)'

# Filter ratings for the specific movie
fight_club_ratings = df[df['movieId'] == movie_title]['rating']

# Plot histogram
plt.hist(fight_club_ratings, bins=5, edgecolor='black')  # You can adjust the number of bins
plt.title(f'Distribution of User Ratings for "{movie_title}"')
plt.xlabel('Rating')
plt.ylabel('Number of Ratings')
plt.show()
```
